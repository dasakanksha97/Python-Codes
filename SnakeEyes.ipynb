```python

import random
count=0

lst_die1 = []
lst_die2 = []

for i in range(0,1000): # trying to make an infinite loop? while true
  die1 = random.randint(1,6)
  die2 = random.randint(1,6)
  print("",i,"Die1: ",die1, "and Die2: ",die2)
  
  lst_die1.append(die1) 
  lst_die2.append(die2)
  count+=1
  if lst_die1[i] == 1 and lst_die2[i] == 1:
    break

print("You've got snake eyes!")
print(f"Number of rolls = {count}")



double_count=0

for one, two in zip(lst_die1, lst_die2):
    if one == two:
        double_count += 1


print("Number of doubles = ", double_count)
      
sum_val1=0

for j in lst_die1:
    sum_val1 += j
avg1 = sum_val1/len(lst_die1)
print(f"Average Roll for Die #1 = {avg1}")


sum_val2 =0

for k in lst_die2:
    sum_val2 += k
avg2 = sum_val2/len(lst_die2)
print(f"Average Roll for Die #2 = {avg2}") 
```
